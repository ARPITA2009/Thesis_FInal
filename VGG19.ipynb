```python
base_path = "/content/breast-histopathology-images/IDC_regular_ps50_idx5/"
folder = listdir(base_path)
print("No. of Patients:",len(folder))
```

```python
total_images = 0
for n in range(len(folder)):
    patient_id = folder[n]
    for c in [0, 1]:
        patient_path = base_path + patient_id
        class_path = patient_path + '/' + str(c) + '/'
        subfiles = listdir(class_path)
        total_images += len(subfiles)
        
print("Total Images in dataset: ", total_images )
```

```python
data = pd.DataFrame(index=np.arange(0, total_images), columns=["patient_id", "path", "target"])

k = 0
for n in range(len(folder)):
    patient_id = folder[n]
    patient_path = base_path + patient_id 
    for c in [0,1]:
        class_path = patient_path + "/" + str(c) + "/"
        subfiles = listdir(class_path)
        for m in range(len(subfiles)):
            image_path = subfiles[m]
            data.iloc[k]["path"] = class_path + image_path
            data.iloc[k]["target"] = c
            data.iloc[k]["patient_id"] = patient_id
            k += 1  

data.head()
```

```python
train_path = 'base_dir/train_dir'
valid_path = 'base_dir/val_dir'
test_path = 'base_dir/test_dir'

num_train_samples = len(df_train)
num_val_samples = len(df_val)
train_batch_size = 100
val_batch_size = 100

train_steps = np.ceil(num_train_samples / train_batch_size)
val_steps = np.ceil(num_val_samples / val_batch_size)
```

```python
IMAGE_SIZE = 75
```

```python
datagen = ImageDataGenerator(rescale = 1.0 / 255,
                             rotation_range = 90,
                             zoom_range = 0.2,
                             horizontal_flip=True,
                             vertical_flip=True)

train_gen = datagen.flow_from_directory(train_path,
                                        target_size=(IMAGE_SIZE,IMAGE_SIZE),
                                        batch_size=train_batch_size,
                                        class_mode='categorical')

val_gen = datagen.flow_from_directory(valid_path,
                                        target_size=(IMAGE_SIZE,IMAGE_SIZE),
                                        batch_size=val_batch_size,
                                        class_mode='categorical')

# Note: shuffle=False causes the test dataset to not be shuffled
test_gen = datagen.flow_from_directory(test_path,
                                        target_size=(IMAGE_SIZE,IMAGE_SIZE),
                                        batch_size=1,
                                        class_mode='categorical',
                                        shuffle=False)
```

```python
import tensorflow as tf
VGG19_model= tf.keras.applications.vgg19.VGG19(include_top=False, weights='imagenet',input_shape=[IMAGE_SIZE,IMAGE_SIZE,3])
VGG19_model.trainable = True
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Flatten,Dense,Dropout,BatchNormalization, MaxPooling2D, AveragePooling2D
  
model2=Sequential()
model2.add(VGG19_model)

model2.add(Conv2D(64, (3, 3), padding = 'same', activation = 'relu', input_shape=[IMAGE_SIZE,IMAGE_SIZE,3]))
model2.add(MaxPooling2D(2, 2))
model2.add(Dropout(0.4))

model2.add(Conv2D(128, (3, 3), padding = 'same', activation = 'relu', input_shape=[IMAGE_SIZE,IMAGE_SIZE,3]))
model2.add(Dropout(0.4))
'''
model2.add(Conv2D(256, (3, 3), padding = 'same', activation = 'relu', input_shape=[IMAGE_SIZE,IMAGE_SIZE,3]))
model2.add(Dropout(0.4))

model2.add(Conv2D(256, (3, 3), padding = 'same', activation = 'relu', input_shape=[IMAGE_SIZE,IMAGE_SIZE,3]))
model2.add(Dropout(0.7))
'''
model2.add(Flatten())
model2.add(BatchNormalization())
model2.add(Dropout(0.35))
model2.add(Dense(2,activation='softmax'))
            
model2.summary()

```

```python
from tensorflow import keras
def ensemble_loss(y_true,y_test):
  bce = keras.losses.BinaryCrossentropy()
  loss_fn1 = bce(y_true,y_test)
  loss = loss_fn1
  mae = keras.losses.MeanAbsoluteError()
  #loss_fn2 = mae(y_true,y_test)
  #loss= 0.7*loss_fn1  + 0.3*loss_fn2
  h = keras.losses.Hinge()
  p = keras.losses.Poisson()
  loss_fn1 = bce(y_true,y_test)
  loss_fn2 = mae(y_true,y_test)
  loss_fn3 = h(y_true,y_test)
  loss_fn4 = p(y_true,y_test)
  #loss= 0.7*loss_fn1 + 0.15*loss_fn3 + 0.15*loss_fn2
  return loss
```

```python
model2.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=ensemble_loss,
              metrics=['accuracy'])

checkpoint = ModelCheckpoint(filepath=os.path.join(MODEL_DIR, "VGG19_model4.h5"), monitor='val_accuracy', verbose=1, 
                             save_best_only=True, mode='max')

reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=3, 
                                   verbose=1, mode='max', min_lr=0.00001)
                              
earlystop_callback = EarlyStopping(monitor='val_loss', min_delta=0.0001, patience=10)

callbacks_list = [checkpoint, reduce_lr, earlystop_callback]

history2 = model2.fit(train_gen, steps_per_epoch=train_steps, 
                    validation_data=val_gen,
                    validation_steps=val_steps,
                    epochs=150, verbose=1,
                   callbacks=callbacks_list)
```

```python
def plot_confusion_matrix(cm, classes,
                         normalize=False,
                         title='Confusion Matrix',
                         cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
        
    print(cm)
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    
```
